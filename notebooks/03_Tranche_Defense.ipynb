## The Simulation

You manage the **Capital Waterfall** of a GPU-backed financing vehicle (SPV).

### The Mechanic: Priority of Payments

Each quarter, revenue flows through a strict hierarchy:

```
üí∞ Revenue (from GPU rentals)
    ‚Üì
1Ô∏è‚É£ Operating Expenses (Electricity) ‚Üê MUST PAY
    ‚Üì

```

**The Trap**: Your asset (H100 GPUs) depreciates ~5% per quarter due to Moore's Law. By Quarter 6, revenue may not cover debt service.

## üìä The Capital Waterfall Explained

### Priority of Payments (Strict Legal Hierarchy):

#### Tier 1: Operating Expenses
**Amount**: ~$20M/quarter  
**Purpose**: Electricity to run GPUs  
**Consequence of Non-Payment**: Immediate shutdown (no revenue)  
**Seniority**: Absolute (even senior debt waits)

#### Tier 2: Senior Debt Interest
**Amount**: $6.25M/quarter (5% APR on $500M)  
**Purpose**: Interest to senior lenders (banks, insurance companies)  
**Consequence of Non-Payment**: Cross-default, asset seizure  
**Seniority**: First lien on assets

#### Tier 3: Mezzanine Debt Interest
**Amount**: $6M/quarter (12% APR on $200M)  
**Purpose**: Interest to mezzanine lenders (private credit funds)  
**Consequence of Non-Payment**: Forced bankruptcy filing  
**Seniority**: Second lien (subordinated to senior)

#### Tier 4: Equity
**Amount**: Whatever remains  
**Purpose**: Returns to equity investors  
**Consequence of Zero**: Dilution, loss of control, wipeout  
**Seniority**: Last (residual claimant)

## üß† Strategy Tips

### Optimal Strategies:

#### 1. **De-Leverage Early** (Conservative)
- Use Quarters 1-3 to pay down mezzanine debt
- Each $50M payment saves $1.5M/quarter in interest
- Goal: Reduce fixed costs before shock hits
- **Risk**: Miss out on dividends if you survive anyway

#### 2. **Extract Dividends Aggressively** (Risky)
- Take money while revenue is high
- Bet you can survive shock with minimal reserves
- Maximize equity value before potential default
- **Risk**: Insufficient cushion for Quarter 5 shock

#### 3. **Balanced Approach** (Moderate)
- One de-leveraging move (Q2 or Q3)
- One dividend extraction (Q1 or Q4)
- Maintain ~$40M in reserves
- **Risk**: Neither maximizes value nor ensures survival

### Key Decision Points:

**Quarter 2-3**: 
- Revenue still strong (~$70M)
- Best time to de-leverage (have excess cash)
- Alternative: Extract dividend before shock

**Quarter 4**: 
- Last chance before shock
- Check reserves: >$50M? Consider dividend
- <$50M? Save every dollar

**Quarter 5**: 
- DeepSeek shock hits (-$30M revenue)
- No good moves, only survival
- Hope you de-leveraged earlier

**Quarter 6-7**: 
- Death spiral begins
- Revenue <$60M, debt service still $12M+
- Burning reserves each quarter
- Cross your fingers

## Game Engine Implementation

The `TrancheGame` class implements the capital waterfall logic:

### Core Components:

1. **Revenue Model**: Base revenue √ó decay factor + volatility
2. **Waterfall Execution**: Strict payment hierarchy
3. **Reserve Management**: Cash buffer for shortfalls
4. **Shock Events**: DeepSeek release at Quarter 5

### Key Variables:
- `debt_senior`, `debt_mezz`, `equity`: Capital stack balances
- `cash`: Reserve buffer for covering shortfalls
- `base_rev`: Starting revenue ($80M)
- `opex`: Fixed operating expenses ($20M)
- `decay`: Quarterly depreciation factor (0.95 = 5% decay)

````python
class TrancheGame:
    def __init__(self):
        self.turn = 1
        self.is_over = False
        self.logs = ["Q1: Operations Started."]
        
        # Capital Stack
        self.debt_senior = 500  # $M, 5% APR
        self.debt_mezz = 200    # $M, 12% APR
        self.equity = 100       # $M, residual claimant
        
        # Operating Variables
        self.cash = 50          # $M reserves
        self.base_rev = 80      # $M per quarter
        self.opex = 20          # $M per quarter (electricity)
        self.decay = 0.95       # 5% quarterly revenue decay (Moore's Law)
        
        self.snapshot = None    # Stores last waterfall result for display

    def next_turn(self, action):
        """Process a quarter: action ‚Üí revenue ‚Üí waterfall ‚Üí check survival."""
        if self.is_over: return
        
        # 1. Action Processing
        if action == "De-Leverage" and self.cash >= 50:
            self.debt_mezz -= 50
            self.cash -= 50
            self.logs.append("üîß Paid down Mezz debt by $50M.")
        elif action == "Dividend" and self.cash >= 20:
            self.cash -= 20
            self.equity -= 10  # Dilution/tax cost
            self.logs.append("üí∏ Extracted $20M dividend (cost $10M equity).")

        # 2. Revenue Generation
        # Base revenue decays, plus random volatility
        rev = max(0, (self.base_rev * (self.decay ** self.turn)) + np.random.normal(0, 5))
        
        # DeepSeek Shock at Q5
        if self.turn == 5: 
            rev -= 30
            self.logs.append("üö® DeepSeek Release! Revenue crashed.")
        
        # 3. Capital Waterfall
        snapshot = {
            "Inflow": rev, 
            "Opex": False, 
            "Senior": False, 
            "Mezz": False, 
            "Equity": 0
        }
        curr = rev  # Remaining cash after each tier
        
        # Tier 1: Operating Expenses (MUST PAY)
        if curr >= self.opex:
            curr -= self.opex
            snapshot["Opex"] = True
        else:
            self.fail("Opex Unpaid - SPV Shutdown")
            return
        
        # Tier 2: Senior Debt Interest
        int_sen = self.debt_senior * 0.05 / 4  # Quarterly interest
        if not self.is_over:
            if curr >= int_sen:
                curr -= int_sen
                snapshot["Senior"] = True
            elif self.cash >= (int_sen - curr):
                # Use reserves to cover shortfall
                self.cash -= (int_sen - curr)
                snapshot["Senior"] = True
                curr = 0
                self.logs.append(f"‚ö†Ô∏è Used reserves to cover senior debt.")
            else:
                self.fail("Senior Default - Lenders Seize Assets")
                return

        # Tier 3: Mezzanine Debt Interest
        int_mezz = self.debt_mezz * 0.12 / 4  # Quarterly interest
        if not self.is_over:
            if curr >= int_mezz:
                curr -= int_mezz
                snapshot["Mezz"] = True
            elif self.cash >= (int_mezz - curr):
                # Use reserves to cover shortfall
                self.cash -= (int_mezz - curr)
                snapshot["Mezz"] = True
                curr = 0
                self.logs.append(f"‚ö†Ô∏è Used reserves to cover mezzanine debt.")
            else:
                self.fail("Mezz Default - Forced Bankruptcy")
                return

        # Tier 4: Equity (Residual)
        if not self.is_over:
            snapshot["Equity"] = curr
            self.cash += curr
            self.snapshot = snapshot
            self.logs.append(f"üåä Rev: ${rev:.1f}M. Equity received: +${curr:.1f}M")
            
        # Check Win Condition
        if self.turn >= 8 and not self.is_over:
            self.is_over = True
            self.logs.append("üèÜ IPO SUCCESS! Exited before maturity.")
            
        if not self.is_over: 
            self.turn += 1

    def fail(self, reason):
        """Trigger game over with failure reason."""
        self.is_over = True
        self.logs.append(f"‚ò†Ô∏è {reason}")
````

## User Interface Setup

Creates an interactive dashboard showing:
- Current quarter and cash reserves
- Visual waterfall with color-coded payment status
- Equity cash flow received this quarter
- Activity log

````python
# UI Setup
td = TrancheGame()
out_td = widgets.Output()

def render_td():
    """Render the waterfall dashboard."""
    with out_td:
        clear_output()
        display(HTML(f"<h3>üóì Quarter: {td.turn} | üí∞ Reserves: ${td.cash:.1f}M</h3>"))
        
        # Waterfall Visualization
        if td.snapshot:
            s = td.snapshot
            
            # Helper function for colored bars
            def bar(label, paid): 
                bg = '#cfc' if paid else '#fcc'  # Green if paid, red if not
                return f"<div style='border:1px solid #ccc; padding:5px; margin:2px; background:{bg}'>{label}</div>"
            
            display(HTML(f"<b>Revenue Inflow: ${s['Inflow']:.1f}M</b>"))
            display(HTML(bar("1. OPEX (Electricity)", s["Opex"])))
            display(HTML(bar("2. Senior Interest", s["Senior"])))
            display(HTML(bar("3. Mezz Interest", s["Mezz"])))
            display(HTML(f"<div style='color:green; padding:5px; margin-top:10px'><b>‚¨á Equity Received: ${s['Equity']:.1f}M</b></div>"))
        
        # Activity Log
        print("\n--- ACTIVITY LOG ---")
        print("\n".join(td.logs[-5:]))
````

## Event Handlers & Game Controls

Sets up the action buttons:

## üìö Lessons from Tranche Defense

### What This Simulation Teaches:

#### 1. High Leverage + Depreciating Assets = Inevitable Default

**The Math Doesn't Work**:
- Traditional infrastructure: Assets appreciate or stay stable
- AI compute: Assets depreciate 50%+ over 3 years
- Fixed debt service + declining revenue = death spiral

**Coverage Ratio Over Time**:
```
Quarter    Revenue    Debt Service    Coverage Ratio
‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
Q1         $80M       $12.25M         6.5x  ‚úÖ
Q3         $72M       $12.25M         5.9x  ‚úÖ
Q5         $40M*      $12.25M         3.3x  ‚ö†Ô∏è
Q7         $60M       $12.25M         4.9x  ‚ö†Ô∏è
Q9         $52M       $12.25M         4.2x  ‚ö†Ô∏è
Q12        $43M       $12.25M         3.5x  ‚ò†Ô∏è
*Includes DeepSeek shock
```

Even without shocks, coverage degrades to dangerous levels by Q12.

#### 2. The Waterfall Structure Protects Debt, Wipes Out Equity

**Equity is the First Loss Absorber**:
- Small revenue decline: Equity returns fall 100%+
- Medium decline: Equity gets zero, burns reserves
- Large decline: Debt defaults, equity wiped out

**Why Lenders Love This Structure**:
- They get paid first
- Fixed coupon (no upside sharing)
- Equity absorbs all volatility
- Assets as collateral if default

**Why Equity Hates This Structure**:
- Last in line for cash flows
- All downside, limited upside
- Residual claim on depreciating assets
- "Picking up pennies in front of a steamroller"

#### 3. Moore's Law Makes Time the Enemy

**Traditional Infrastructure**:
- Time ‚Üí Assets mature ‚Üí Higher cash flows (pricing power)
- Example: Toll road traffic grows with population

**AI Compute Infrastructure**:
- Time ‚Üí Assets obsolete ‚Üí Lower cash flows (price competition)
- Example: H100 rental $4/hr (2023) ‚Üí $2/hr (2025) ‚Üí $0.50/hr (2026)

**The 5% Quarterly Decay**:
This models real GPU depreciation:
- Year 1: 100% of initial revenue
- Year 2: 77% of initial revenue
- Year 3: 54% of initial revenue
- Year 4: 37% of initial revenue

By Year 4, cash flows cannot support Year 1 leverage levels.

#### 4. Shocks Accelerate the Inevitable

**Without DeepSeek Shock**:
- Survive to Q8 (barely)
- Equity returns compress to near-zero
- Default likely by Q10-12

**With DeepSeek Shock**:
- Default probability 50%+ by Q6
- Reserves critical for survival
- De-leveraging in Q2-3 is make-or-break

The shock doesn't **cause** failure‚Äîit **reveals** the fragility that was always there.

#### 5. Optimal Strategy Requires Foresight

**Naive Approach** ("Maximize Short-Term Cash"):
- Extract dividends in Q1-3
- Build reserves for shock
- **Result**: Insufficient buffer, default at Q5-6

**Survival Approach** ("Minimize Fixed Costs"):
- De-leverage in Q2 and Q3 (reduce mezz debt by $100M)
- Lower quarterly interest from $12.25M to $9.25M
- **Result**: Survive Q5 shock, reach Q8 IPO

**Greedy Approach** ("Take the Money and Run"):
- Extract max dividends before shock
- Accept high default risk
- **Result**: Maximize equity value in success states, total loss in failure

The "right" strategy depends on:
- Your risk tolerance
- Beliefs about shock timing/severity
- Other investors in the capital structure

## üí° Systemic Implications

### Why This Matters for Financial Stability:

#### 1. **Misclassification of Asset Class**

**Lenders Think**:
"We're lending against infrastructure. It's like data centers, which have 20-year cash flows and stable returns."

**Reality**:
"We're lending against consumer electronics. It's like financing iPhone inventory‚Äîobsolete in 2 years."

This classification error leads to:
- Inappropriate leverage (80% LTC vs. 50% for consumer tech)
- Inappropriate loan terms (5-year maturity vs. 2-year for tech)
- Inappropriate covenants (loose vs. tight for volatile assets)

#### 2. **Mark-to-Market vs. Hold-to-Maturity Accounting**

Private credit funds hold these loans at "amortized cost" (book value), not market value:

**Lender's Books (Year 2)**:
- Loan Balance: $230M (original $240M - $10M paydown)
- Mark: Par (100% of face value)
- **Reported Asset**: $230M ‚úÖ

**Market Reality (Year 2)**:
- Loan Value: ~$150M (65 cents on dollar, priced for default risk)
- **True Economic Value**: $150M ‚ò†Ô∏è

**Hidden Loss**: $80M (35% of original loan)

This accounting treatment hides losses until default, at which point:
- Fund NAV collapses overnight
- LPs panic and request redemptions
- Fund forced to sell other loans at fire-sale prices
- **Contagion to unrelated borrowers**

#### 3. **Concentration Risk Across the Industry**
